```python
#Setting seprate columns of Date, Month & Year from Date_of_Journey column 

merge_df["Date"] = merge_df["Date_of_Journey"].str.split('/').str[0]
merge_df["Month"] = merge_df["Date_of_Journey"].str.split('/').str[1]
merge_df["Year"] = merge_df["Date_of_Journey"].str.split('/').str[2]
```

```python
#merge_df["Date"] = merge_df["Date"].apply(lambda x :int(x)).astype('int32')

merge_df["Date"] = merge_df["Date"].astype(int)
merge_df["Month"] = merge_df["Month"].astype(int)
merge_df["Year"] = merge_df["Year"].astype(int)
```
